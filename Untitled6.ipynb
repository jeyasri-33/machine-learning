```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Salary_Data.csv")
df
```

```python
independent=df[["YearsExperience"]]
independent
```

```python
dependent=df[["Salary"]]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.2, random_state=0)
```

```python
import numpy as np
from sklearn import linear_model
regressor = linear_model.RidgeCV(alphas=np.logspace(-6, 6, 13))
regressor.fit(x_train, y_train)


weight = regressor.coef_
print("weight of the model={}", format (weight))
bais = regressor.intercept_
print("bais of the model ={}", format (bais))
y_pred = regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
print("r_score=", r_score)
import numpy as np
YearsExperience=int(input("enter the value:"))
Salary=int(input("enter ther value:"))





future_prediction = regressor.predict([[YearsExperience,Salary]])
print("future_prediction = {}", format(future_prediction))
```

```python
import pickle
filename="finalModel.sav"
pickle.dump(regressor,open(filename,'wb'))
```
